```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
class PuzzleLinear:
    def __init__(self, rows, cols, image_path):
        self.rows = rows
        self.cols = cols
        self.image = np.array(Image.open(image_path).resize((cols * 100, rows * 100)))
        self.grid = self.create_grid()

    def create_grid(self):
        x = np.linspace(0, 1, self.cols + 1)
        y = np.linspace(0, 1, self.rows + 1)
        return np.meshgrid(x, y)

    def plot_puzzle(self):
        x, y = self.grid
        plt.figure(figsize=(8, 8))
        plt.imshow(self.image, extent=(0, 1, 0, 1), aspect="auto")

        # Choose random equidistant points for vertical lines
        x_points = np.linspace(0, 1, self.cols + 1)
        x_offsets = np.random.uniform(-0.1, 0.1, size=self.cols + 1)  # Adjust 0.1 for more/less variation
        x_points = np.clip(x_points + x_offsets, 0, 1)  # Ensure points stay within the image

        # Choose random equidistant points for horizontal lines
        y_points = np.linspace(0, 1, self.rows + 1)
        y_offsets = np.random.uniform(-0.1, 0.1, size=self.rows + 1)  # Adjust 0.1 for more/less variation
        y_points = np.clip(y_points + y_offsets, 0, 1)  # Ensure points stay within the image

        # Draw vertical lines
        for x_start in x_points:
            plt.plot([x_start, x_start], [0, 1], color="black")

        # Draw horizontal lines
        for y_start in y_points:
            plt.plot([0, 1], [y_start, y_start], color="black")

        plt.axis("equal")
        plt.axis("off")
        plt.show()
```

```python
image_path = 'Images/Rei.jpg'
puzzle = PuzzleLinear(rows=6, cols=6, image_path=image_path)
puzzle.plot_puzzle()
```
