```python
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('Datasets/RockData.csv')

df_selected = df[['popularity', 'danceability']].values
```

```python
# here i scale data for better results
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df_selected)

dbscan = DBSCAN(eps=0.2, min_samples=20)
clusters = dbscan.fit_predict(df_scaled)
```

```python
df['cluster'] = clusters

# then visualize the results
plt.figure(figsize=(10, 6))
plt.scatter(df['popularity'], df['danceability'], c=df['cluster'], cmap='viridis', marker='o', edgecolor='k')
plt.title('DBSCAN Clustering of Popularity vs Danceability')
plt.xlabel('Popularity')
plt.ylabel('Danceability')
plt.colorbar(label='Cluster Label')
plt.show()
```
