```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
f = lambda x :  np.e ** x + 4 * x * x

exact_derivative = lambda x: np.e ** x + 8 * x

forward_diff = lambda f, x, h: (f(x + h) - f(x)) / h

backward_diff = lambda f, x, h: (f(x) - f(x - h)) / h

central_diff = lambda f, x, h: (f(x + h) - f(x - h)) / (2 * h)
```

```python
x_values = np.linspace(5, 10, 100)
```

```python
# h_input = input("Enter a step size h or press enter for h to be 0.1 : ")
# h = float(h_input) if h_input else 0.1
h = 0.1
```

```python
forward_approx = forward_diff(f, x_values, h)
backward_approx = backward_diff(f, x_values, h)
central_approx = central_diff(f, x_values, h)
exact_values = exact_derivative(x_values)
```

```python
plt.figure(figsize=(10, 6))

plt.plot(x_values, exact_values, label="Exact Derivative", color="black", linewidth=2)

plt.plot(x_values, forward_approx, label="Forward Difference", linestyle='--', color="blue", linewidth=1.5)
plt.plot(x_values, backward_approx, label="Backward Difference", linestyle='-.', color="red", linewidth=1.5)
plt.plot(x_values, central_approx, label="Central Difference", linestyle=':', color="green", linewidth=1.5)

plt.title("Comparison of Finite Difference Methods with Exact Derivative")
plt.xlabel("x")
plt.ylabel("Derivative")
plt.legend()
plt.grid(True)

plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(x_values, forward_approx - exact_values, label="Forward Difference Error", linestyle='--', color="blue")
plt.plot(x_values, backward_approx - exact_values, label="Backward Difference Error", linestyle='-.', color="red")
plt.plot(x_values, central_approx - exact_values, label="Central Difference Error", linestyle=':', color="green")

plt.title("Error Comparison Between Finite Difference Methods")
plt.xlabel("x")
plt.ylabel("Error")
plt.legend()
plt.grid(True)
plt.show()
```
