```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def lagrange_interpolation(x, x_vals, y_vals):
    result = 0
    n = len(x_vals)
    for i in range(n): 
        l = y_vals[i] 
        for j in range(n):
            if i != j:
                l *= (x - x_vals[j]) / (x_vals[i] - x_vals[j])
        result += l
    return result
```

```python
points = np.array([(-16.32, 87.09), (-7.18, 53.7), (-3, 23.73),(16.32, -87.09), (34.55, 136)])
x_values, y_values = points[:, 0], points[:, 1]

x_vals = np.linspace(min(x_values) - 1, max(x_values) + 1, 100)
y_vals = [lagrange_interpolation(x, x_values, y_values) for x in x_vals]
```

```python
plt.figure(figsize=(10, 8))
plt.plot(x_vals, y_vals, label="This is an interpolaterd", color='red')
plt.scatter(x_values, y_values, color='brown', label='Data Points')
plt.xlabel("x")
plt.ylabel("y")
plt.title("Lagrange Polynomial Interpolation")
plt.legend()
plt.grid(True)
plt.show()
```
