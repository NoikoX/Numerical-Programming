```python
import cv2
import matplotlib
import numpy as np
from matplotlib import pyplot as plt
import matplotlib.image as mpimg
```

```python
image = cv2.imread("Pawawo.jpg", cv2.IMREAD_GRAYSCALE)

img = mpimg.imread("Pawawo.jpg")
plt.imshow(img)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.imshow(image, cmap='gray')
plt.title("Original Image - Pawawo(Gray)")
plt.axis('off')
plt.show()
```

```python
gaussian_blur = cv2.GaussianBlur(image, (3, 3), 0)

plt.figure(figsize=(8, 6))
plt.imshow(gaussian_blur, cmap='gray')
plt.title("Gaussian Blur")
plt.axis('off')
plt.show()
```

```python
sobel_x = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=3)

plt.figure(figsize=(8, 6))
plt.imshow(sobel_x, cmap='gray')
plt.title("Sobel X")
plt.axis('off')
plt.show()
```

```python
sobel_y = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=3)

plt.figure(figsize=(8, 6))
plt.imshow(sobel_y, cmap='gray')
plt.title("Sobel Y")
plt.axis('off')
plt.show()
```

```python
sobel_both = cv2.magnitude(sobel_x, sobel_y)

plt.figure(figsize=(8, 6))
plt.imshow(sobel_both, cmap='gray')
plt.title("Sobel Combined")
plt.axis('off')
plt.show()
```

```python
scharr_x = cv2.Scharr(image, cv2.CV_64F, 1, 0)

plt.figure(figsize=(8, 6))
plt.imshow(scharr_x)
plt.title("Scharr X")
plt.axis('off')
plt.show()
```

```python
scharr_y = cv2.Scharr(image, cv2.CV_64F, 0, 1)

plt.figure(figsize=(8, 6))
plt.imshow(scharr_y, cmap='gray')
plt.title("Scharr Y")
plt.axis('off')
plt.show()
```

```python
scharr_both = cv2.magnitude(scharr_x, scharr_y)

plt.figure(figsize=(8, 6))
plt.imshow(scharr_both, cmap='gray')
plt.title("Scharr Combined")
plt.axis('off')
plt.show()
```

```python
laplacian = cv2.Laplacian(image, cv2.CV_64F)

plt.figure(figsize=(8, 6))
plt.imshow(laplacian, cmap='gray')
plt.title("Laplacian")
plt.axis('off')
plt.show()
```
